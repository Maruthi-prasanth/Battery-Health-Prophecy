```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data from .mat file
loaded_data = scipy.io.loadmat('B0018.mat')
```

```python
data=loaded_data['B0018']
```

```python
#All Charge Cycles
c=[]
for row in data[0][0][0][0]:
    if(row[0]==['charge']):
        charge=row[-1]
        c.extend(charge)
print(c)
```

```python
#Appending Voltage and current measured in lists 
f_item=[]
fc_item=[]
ft_item=[]
for c_item in c:
    for v_item in c_item:
        if len(v_item)>0:
            f_item.append(v_item[0])
            fc_item.append(v_item[1])
            ft_item.append(v_item[5])
```

```python
#Voltage list elements for every cycle 
i=0
for roww in f_item:
    for ele in roww:
        print(i,ele,end=" ")
    i=i+1
    print()
```

```python
#Voltage Time Graph 
for i in range(0,134):
    plt.plot(f_item[i][0])
plt.xlabel('Time(s)')
plt.ylabel('Voltage(V)')
plt.ylim(3.0,4.5)
plt.show()
```

```python
#Current Time Graph 
for i in range(0,134):
    plt.plot(fc_item[i][0])
plt.xlabel('Time(s)')
plt.ylabel('Current(A)')
plt.ylim(0,3.0)
plt.show()
```

```python
#HI 1
#voltage values 
volts = []
for roww in f_item:
    for ele in roww:
        for volt in ele:
            if volt > 3.9000 and volt <= 4.0000:
                volts.append(volt)
                break
print(volts)
```

```python
#indexes of that voltage values 
indexes = []
for roww in f_item:
    for ele in roww:
        n = 0
        for volt in ele:
            if volt > 3.9000 and volt <= 4.1000:
                indexes.append(n + 1)
                break
            n = n + 1
print(indexes)
```

```python
#time values corresponding to 3.9 indexes
times = []
idx = 0
for roww in ft_item:
    for ele in roww:
        n = 1
        for time in ele:
            if idx < len(indexes) and n == indexes[idx]:
                times.append(time)
                break 
            n = n + 1
        idx = idx + 1

print(times)
```

```python
#indexes of that voltage values 
indexes1 = []
for roww in f_item:
    for ele in roww:
        n = 0
        for volt in ele:
            if volt > 4.2000 and volt <= 4.3000:
                indexes1.append(n + 1)
                break
            n = n + 1
print(indexes1)
```

```python
#time values corresponding to 4.2 indexes
times1 = []
idx = 0
for roww in ft_item:
    for ele in roww:
        n = 1
        for time in ele:
            if idx < len(indexes1) and n == indexes1[idx]:
                times1.append(time)
                break 
            n = n + 1
        idx = idx + 1

print(times1)
```

```python
#Difference between 4.2index time value and 3.9index time value 
diff_list=[]
min_l=min(len(times),len(times1))
for i in range(min_l):
    diff_list.append(times1[i]-times[i])

print(diff_list)
```

```python
plt.xlabel("Cycle Number")
plt.ylabel("Time difference")
plt.title("HI 1 Before Data Cleaning(B0018)")
plt.plot(diff_list)
```

```python
#data cleaning on HI1
for i in range(len(diff_list)):
    if diff_list[i] < 900.0:
        diff_list[i] = 2450.0
print(diff_list) 
```

```python
plt.xlabel("Cycle Number")
plt.ylabel("Time difference")
plt.title("HI 1 for B0018 Battery Dataset")
plt.plot(diff_list)
plt.ylim(0,3000)
```

```python
### HI 2
#Time value after 500 seconds 
time_500=[]
for i in range(len(times)):
    updated=times[i]+500
    time_500.append(updated)

print(time_500)
```

```python
#time indexes after 500 seconds 
n_indexes = []
idx = 0
for roww in ft_item:
    for ele in roww:
        n = 1
        for time in ele:
            if idx < len(time_500) and time >= time_500[idx]:
                n_indexes.append(n)
                break
            n = n + 1
        idx = idx + 1

print(n_indexes)
```

```python
#voltage corresponding to indexes after 500 seconds at 3.9
voltss = []
idx = 0
for roww in f_item:
    for ele in roww:
        n = 1
        for time in ele:
            if idx < len(n_indexes) and n == n_indexes[idx]:
                voltss.append(time)
                break 
            n = n + 1
        idx = idx + 1

print(voltss)
```

```python
#Difference between 3.9index voltage value and after 500 seconds 
diff_list1=[]
min_l1=min(len(voltss),len(volts))
for i in range(min_l1):
    diff_list1.append(voltss[i]-volts[i])

print(diff_list1)
```

```python
plt.xlabel("Cycle Number")
plt.ylabel("Voltage difference")
plt.title("HI 2 Before Data Cleaning(B0018)")
plt.plot(diff_list1)
```

```python
 #data cleaning on HI2
for i in range(len(diff_list1)):
    if diff_list1[i] > 0.15:
        diff_list1[i] = 0.07
print(diff_list1) 
```

```python
plt.xlabel("Cycle Number")
plt.ylabel("Voltage difference")
plt.title("HI 2 for B0018 Battery Dataset")
plt.plot(diff_list1)
plt.ylim(0,0.3)
```

```python
### HI 3
#current values
currs = []
for roww in fc_item:
    for ele in roww:
        for volt in ele:
            if volt>1.4500 and volt <= 1.5000:
                currs.append(volt)
                break
print(currs)
```

```python
#indexes of that current values 
c_indexes = []
for roww in fc_item:
    for ele in roww:
        n = 0
        for volt in ele:
            if volt > 1.4000 and volt < 1.5000:
                c_indexes.append(n + 1)
                break
            n = n + 1
print(c_indexes)
```

```python
#time values corresponding to current indexes
c_times = []
idx1 = 0
for roww in ft_item:
    for ele in roww:
        n = 1
        for time in ele:
            if idx1 < len(c_indexes) and n == c_indexes[idx1]:
                c_times.append(time)
                break 
            n = n + 1
        idx1 = idx1 + 1

print(c_times)
```

```python
#Time value after 1000 seconds 
time_1000=[]
for i in range(len(c_times)):
    updated1=c_times[i]+1000
    time_1000.append(updated1)

print(time_1000)
```

```python
#time indexes after 1000 seconds 
c_indexes = []
idx2 = 0
for roww in ft_item:
    for ele in roww:
        n = 0
        for time in ele:
            if idx2 < len(time_1000) and time >= time_1000[idx2]:
                c_indexes.append(n)
                break
            n = n + 1
        idx2 = idx2 + 1

print(c_indexes)
```

```python
#current values corresponding to indexes after 1000 seconds
currs1 = []
idx3 = 0
for roww in fc_item:
    for ele in roww:
        n = 1
        for time in ele:
            if idx3 < len(c_indexes) and n == c_indexes[idx3]:
                currs1.append(time)
                break 
            n = n + 1
        idx3 = idx3 + 1

print(currs1)
```

```python
diff_list2=[]
min_l2=min(len(currs1),len(currs))
for i in range(min_l2):
    diff_list2.append(currs[i]-currs1[i])

print(diff_list2)
```

```python
#HI3 graph
plt.xlabel("Cycle Number")
plt.ylabel("Current(A)")
plt.title("HI 3 Before Data Cleaning(B0018)")
plt.plot(diff_list2)
plt.ylim(-0.2,1.5)
```

```python

#data cleaning on HI3
for i in range(len(diff_list2)):
    if diff_list2[i] > 1.0 :
        diff_list2[i] = 0.9
print(diff_list2) 
```

```python

#data cleaning on HI3
for i in range(len(diff_list2)):
    if diff_list2[i] < 0.6 :
        diff_list2[i] = 0.85
print(diff_list2) 
```

```python
plt.xlabel("Cycle Number")
plt.ylabel("Current(A)")
plt.title("HI 3 for B0018 Battery Dataset")
plt.plot(diff_list2)
plt.ylim(0,1.5)
```

```python
from scipy.stats import pearsonr, spearmanr
```

```python
#capacity values in our data set
capacity = []

for row in data[0][0][0][0]:
    if row[0] == ['discharge']:
        capacity.append(row[-1][-1][-1][-1][0][0])
        
print(capacity)
```

```python
#correlation of HI1
minn=min(len(capacity),len(diff_list))
capacity=capacity[:minn]
diff_list=diff_list[:minn]
pearson_corr1,_=pearsonr(capacity,diff_list)
spearman_corr1,_=spearmanr(capacity,diff_list)
print(pearson_corr1)
print(spearman_corr1)
```

```python
#correlation on HI2
minn1=min(len(capacity),len(diff_list1))
capacity=capacity[:minn1]
diff_list1=diff_list1[:minn1]
pearson_corr2,_=pearsonr(capacity,diff_list1)
spearman_corr2,_=spearmanr(capacity,diff_list1)
print(pearson_corr2)
print(spearman_corr2)
```

```python
#Correlation on HI3
minn2=min(len(capacity),len(diff_list2))
capacity=capacity[:minn2]
diff_list2=diff_list2[:minn2]
pearson_corr3,_=pearsonr(capacity,diff_list2)
spearman_corr3,_=spearmanr(capacity,diff_list2)
print(pearson_corr3)
print(spearman_corr3)
```
